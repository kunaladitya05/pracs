Aim: Matrix operation and Differentiation of vector and matrix.

```python
import numpy as np
import sympy as sp

A = np.array([[1, 2],
              [3, 4]])

B = np.array([[5, 6],
              [7, 8]])

print("Matrix A:")
print(A)

print("\nMatrix B:")
print(B)

print("\nAddition (A + B):")
print(A + B)

print("\nSubtraction (A - B):")
print(A - B)

print("\nMatrix Multiplication (A × B):")
print(np.matmul(A, B))

print("\nElement-wise Multiplication:")
print(A * B)

print("\nTranspose of A:")
print(A.T)

print("\nInverse of A:")
print(np.linalg.inv(A))

x = sp.symbols('x')

vector = sp.Matrix([
    x**2,
    sp.sin(x),
    sp.exp(x)
])

print("\nVector Function:")
sp.pprint(vector)

vector_diff = vector.diff(x)

print("\nDerivative of Vector:")
sp.pprint(vector_diff)

matrix = sp.Matrix([
    [x**2, sp.sin(x)],
    [sp.exp(x), x**3]
])

print("\nMatrix Function:")
sp.pprint(matrix)

matrix_diff = matrix.diff(x)

print("\nDerivative of Matrix:")
sp.pprint(matrix_diff)
```

Aim: Integration of a vector and matrix

```python
import numpy as np

x = np.array([0, 1, 2, 3, 4])

vector = np.array([1, 4, 9, 16, 25])

matrix = np.array([
    [1, 4, 9, 16, 25],
    [2, 5, 10, 17, 26]
])

vector_integral = np.trapezoid(vector, x)

matrix_integral = np.trapezoid(matrix, x, axis=1)

print("Vector Integral:", vector_integral)
print("Matrix Integral:", matrix_integral)
```

Aim: Simplex algorithm and Duality

```python
import numpy as np
from scipy.optimize import linprog

c = [-3, -5]

A = [[1, 2],
     [3, 2]]

b = [8, 12]

result = linprog(c, A_ub=A, b_ub=b, method='simplex')

print("Optimal Solution:")
print("x =", result.x[0])
print("y =", result.x[1])

print("\nMaximum Value of Z:")
print(-result.fun)

print("\nDual Variables:")
print(result.slack)
```

Aim: Implementation of Newton's Method

```python
from scipy.optimize import minimize

def f(x):
    return (x[0] - 3)**2 + (x[1] - 2)**2

x0 = [0, 0]

result = minimize(f, x0, method='BFGS')

print("Optimal Solution:", result.x)
print("Minimum Value:", result.fun)
print("Iterations:", result.nit)
```

Aim: Implementation of Secant method

```python
def f(x):
    return x**3 - x - 2

x0 = 1
x1 = 2

tolerance = 0.0001
max_iter = 100

for i in range(max_iter):

    x2 = x1 - (f(x1) * (x1 - x0)) / (f(x1) - f(x0))

    if abs(x2 - x1) < tolerance:
        print("Root =", round(x2, 4))
        break

    x0 = x1
    x1 = x2
```

Aim: Implementation of Lagrange multiplier method

```python
import sympy as sp

x, y, lam = sp.symbols('x y lam')

f = x**2 + y**2

g = x + y - 1

L = f - lam * g

eq1 = sp.diff(L, x)
eq2 = sp.diff(L, y)
eq3 = sp.diff(L, lam)

solution = sp.solve([eq1, eq2, eq3], (x, y, lam))

print("Solution (x, y, lambda):")
print(solution)
```

Aim: Implementation of KKT theorem

```python
import sympy as sp

x, y, lam = sp.symbols('x y lam')

f = x**2 + y**2

g = x + y - 1

L = f + lam * g

eq1 = sp.diff(L, x)   
eq2 = sp.diff(L, y)  

eq3 = lam * g

solutions = sp.solve([eq1, eq2, eq3], (x, y, lam), dict=True)

print("KKT Solutions:")
for sol in solutions:
    print(sol)
```

Aim: Implementation of BFGS method

```python
import numpy as np
from scipy.optimize import minimize

def f(x):
    return (1 - x[0])**2 + 100 * (x[1] - x[0]**2)**2

x0 = np.array([2.0, 2.0])

result = minimize(f, x0, method='BFGS', options={'disp': True})

print("\nOptimal Solution:")
print("x =", result.x)
print("Minimum value =", result.fun)
```

Aim: Particle Swarm Optimization Algorithm

```python
import numpy as np

def fitness(x):
    return 10 * len(x) + sum(xi**2 - 10 * np.cos(2 * np.pi * xi) for xi in x)

num_particles = 30
dimensions = 2
iterations = 100

w = 0.7   
c1 = 1.5    
c2 = 1.5     

particles = np.random.uniform(-5, 5, (num_particles, dimensions))
velocities = np.random.uniform(-1, 1, (num_particles, dimensions))

pbest = particles.copy()
pbest_score = np.array([fitness(p) for p in particles])

gbest = pbest[np.argmin(pbest_score)]
gbest_score = min(pbest_score)

for _ in range(iterations):
    for i in range(num_particles):
        r1, r2 = np.random.rand(dimensions), np.random.rand(dimensions)

        velocities[i] = (w * velocities[i]
                         + c1 * r1 * (pbest[i] - particles[i])
                         + c2 * r2 * (gbest - particles[i]))

        particles[i] += velocities[i]

        score = fitness(particles[i])

        if score < pbest_score[i]:
            pbest[i] = particles[i]
            pbest_score[i] = score

    gbest = pbest[np.argmin(pbest_score)]
    gbest_score = min(pbest_score)

print("Optimal Solution:", gbest)
print("Best Fitness Value:", gbest_score)
```

Aim: Flower Pollination Algorithm

```python
import numpy as np
import math

def levy_flight(beta=1.5):
    sigma = (math.gamma(1 + beta) * math.sin(math.pi * beta / 2) /
             (math.gamma((1 + beta) / 2) * beta * 2 ** ((beta - 1) / 2))) ** (1 / beta)

    u = np.random.normal(0, sigma)
    v = np.random.normal(0, 1)

    step = u / (abs(v) ** (1 / beta))
    return step


n = 30          
dim = 2       
max_iter = 100

pa = 0.8       

X = np.random.uniform(-5, 5, (n, dim))

F = np.array([fitness(x) for x in X])

best_idx = np.argmin(F)
best = X[best_idx].copy()
best_score = F[best_idx]

for t in range(max_iter):
    for i in range(n):

        if np.random.rand() < pa:
            L = levy_flight()
            X[i] = X[i] + L * (X[i] - best)
        else:
            j, k = np.random.randint(0, n, 2)
            epsilon = np.random.rand()
            X[i] = X[i] + epsilon * (X[j] - X[k])

        F[i] = fitness(X[i])

        if F[i] < best_score:
            best = X[i].copy()
            best_score = F[i]

print("Best Solution:", best)
print("Best Fitness:", best_score)
```
